## II. Brute-force approach to the "Knapsack problem"

```python
# ==================================================== #
# ==============BRUTE FORCE APPROACH================== #
# ==================================================== #

# Importing the necessary libraries
import itertools
import time


# Defining functions for main()
def get_weight(combo):
    '''Returns the associated weight of a combination'''
    return sum(item[1] for item in combo)


def get_value(combo):
    '''Returns the associated weight of a combination'''
    return sum(item[2] for item in combo)


def print_stats(best_solution, optimal_value, computations, total_time):
    '''Print the results and brief evaluation statistics for the search'''

    print("\nBest solution:")
    for item in best_solution:
        print(f"*{item[0].capitalize()}")
        
    print(f"\n===>Best solution value: {get_value(best_solution):,}")
    if get_value(best_solution) == optimal_value:
        print(f"\n===>True optimal solution ({optimal_value:,}) was discovered")
    
    print(f"\nTotal iterations: {computations:,}")
    print(f"Total time taken: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")
    
    
def main ():
    # Defining global variables
    global computations
    global best_solution
    
    # Timing counter for evaluating performance
    start_time = time.perf_counter()

    # Generating combinations in the range: 1-->n (where n= no. available items)
    for i in range(1,len(chosen_item_list)+1):
        for combo in  itertools.combinations(chosen_item_list,i):
            computations += 1
            # Printing iterations in multiples of 50,000 for larger calculations
            if computations % 50_000 == 0:
                print(f"Iteration: {computations:>{pad},}/{max_iterations:,} ({computations/max_iterations*100:.2f}%)", end='\r')

            # Getting the weight and value of the current combination
            current_weight = get_weight(combo)
            current_value = get_value(combo)

            # Setting/updating the best solution based the value of the current combination
            if (current_weight <= max_weight) and (current_value > get_value(best_solution)):
                best_solution = combo
                continue

    # Printing the final iteration log
    print(f"Iteration: {computations:>{pad},}/{max_iterations:,} ({computations/max_iterations*100:.2f}%)")

    # Calculating the total time elapsed
    end_time = time.perf_counter()
    total_time = end_time - start_time

    # Print a brief report for the search
    print_stats(best_solution, optimal_value, computations, total_time)

# ================================================================== #
# ================================================================== #

# Choose the item package for the search
chosen_item_list, max_weight, optimal_value = large_items_package

# Set globals for the brute force search
best_solution = (('dummy', 0, 0),)
computations = 0
max_iterations = 2**(len(chosen_item_list)) - 1
pad = len(f"{max_iterations:,}")

# ================================================================== #
# ================================================================== #

# Running main() if executed directly
if __name__ == "__main__":
    main()
```

## III. Evolutionary algorithm based search for the "Knapsack problem"

```python
# ==================================================== #
# =========EVOLUTIONARY ALGORITHM APPROACH============ #
# ==================================================== #

# Importing the necessary libraries
import random
import time


# Defining functions for main()
def get_weight(individual):
    '''Returns the associated weight of an individual'''
    return sum(chosen_item_list[i][1] for i in range(len(individual)) if individual[i] == 1)


def get_value(individual):
    '''Returns the associated value of an individual'''
    return sum(chosen_item_list[i][2] for i in range(len(individual)) if individual[i] == 1)


def initialize_parents(parents_for_next_generation, max_weight):
    '''Returns a list of valid, randomly generated individuals (parents) for the 1st run of the algorithm'''
    all_parents = []
    for x in range(parents_for_next_generation):
        while True:
            parent = [random.randint(0,1) for x in range(len(chosen_item_list))]
            weight = get_weight(parent)
            if weight <= max_weight: # Accepting an individual if it satisfies the max weight criterion
                all_parents.append(parent)
                break
    return all_parents


def generate_child(*parents, mutation_rate=0.5):
    '''Return a child from a supplied population of parents via random recombination.
    Supports single point random mutations based on a mutation rate (default=0.5)'''
    child = [random.choice(parents)[x] for x in range(len(parents[0]))]
    if random.random() < mutation_rate:
        change_index = random.randint(0, len(child)-1)
        child[change_index] = 0 if child[change_index] == 1 else 1
    return child

    
def get_items(individual):
    '''Returns a list of items included in an individual'''
    return [chosen_item_list[i][0] for i in range(len(individual)) if individual[i] == 1]


def solution_found(current_value, optimal_value):
    '''Returns True is the optimal solution has been found'''
    if current_value == optimal_value:
        return True
    return False


def print_stats(best_individual, optimal_value, current_generation, total_time):
    '''Print the results and brief evaluation statistics for the search'''
    
    print("\nBest solution:")
    for item in get_items(best_individual):
        print(f"*{item}")
    print(f"\n===>Best solution value: {get_value(best_individual):,}")

    if best_individual_value == optimal_value:
        print(f"\n****SEARCH SUCCESSFUL***\nTrue optimal solution ({optimal_value:,}) was discovered")
    else:
        print(f"\n****SEARCH FAILED****\nTrue optimal solution ({optimal_value:,}) was NOT discovered")
    print(f"\nTotal no. of generations: {current_generation+1}")
    print(f"Total time taken: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")


def plot_search_progress(data, total_generations=None, type='value_per_gen'):
    '''Plot the progression of the search
    type could be either "value_per_gen" or "one_hot_encoding" '''
    import matplotlib.pyplot as plt
    import numpy as np
    
    _data = data.copy() # Copying data to avoid side-effects
    
    if type == 'value_per_gen':
        plt.style.use('seaborn-darkgrid')
        fig, ax = plt.subplots(figsize=(8,4),dpi=200)
        ax.plot([x[0] for x in _data], [x[1] for x in _data], label='Best individual value')
        ax.axhline(optimal_value, c='red', lw=2, label='Optimal value')
        ax.legend(frameon=True, facecolor='white', shadow=True)
        ax.set_xlabel("Generations")
        ax.set_ylabel("Value (individual fitness)")
        ax.set_title("Evolutionary algorithm based optimation of the Knapsack problem", pad=20)
    
    elif type == 'one_hot_encoding':
        _data.append(np.zeros(26)) # Using a numpy array to avoid a ragged array 
        _data.append([3 if x == 1 else 0 for x in _data[-2]])
        plt.style.use('seaborn-dark')
        fig, ax = plt.subplots(figsize=(8,8), dpi=200)
        ax.imshow(_data, cmap=plt.get_cmap('Blues'))
        ax.set_ylabel("Generations", fontsize=9)
        ax.set_xlabel("One-hot encoding scheme", fontsize=9)
        ax.set_xticklabels([])
        plt.yticks(fontsize=8)
        for x in range(0, len(_data)-2):
            ax.axhline(x, c='white', lw=1, alpha=0.5)
        
    
    
def main(*,print_report=True, plot_results=True, return_gen=False):
    # Defining global variables
    global best_individual_value
    
    # Timing counter for evaluating performance
    start_time = time.perf_counter()
    
    # Intializing initial parents
    all_parents = initialize_parents(parents_for_next_generation, max_weight)
    
    # Generating lists for logging best individual values for each generation
    best_individual_value_per_generation = []
    best_individual_per_generation = []
    
    # Running the evolutionary algorithm for the specified generations
    for current_generation in range(generations):
        # Printing the current generation number
        if print_report:
            print(f"Generation #{current_generation+1}", end='\r')
        
        # Checking if the optimal solution has been found
        if solution_found(best_individual_value, optimal_value):
            break

        # Including all initial parents as individuals of the current generation
        children = all_parents.copy()

        # Generating children for the current generation
        for j in range(progeny_per_generation):
            child = generate_child(*all_parents, mutation_rate=mutation_rate)
            child_weight = get_weight(child)

            # Rejecting a child if it violates the max weight criterion
            if child_weight > max_weight:
                continue
            else:
                children.append(child)

        # Iterating over the individuals of the current generation
        for child in children:
            # Setting/updating the best solution based on the child's value
            child_value = get_value(child)
            if child_value > best_individual_value:
                best_individual = child
                best_individual_value = child_value

        # Selecting the best individuals from the current as parents for the next generation
        all_parents = sorted(children, key=get_value, reverse=True)[:parents_for_next_generation]
        
        # Logging the generation and the best individual's value
        best_individual_value_per_generation.append((current_generation+1, get_value(all_parents[0])))
        best_individual_per_generation.append(all_parents[0])

    # Calculating the total time elapsed
    end_time = time.perf_counter()
    total_time = end_time - start_time
    
    # Print a report for the search
    if print_report:
        print_stats(best_individual,optimal_value,current_generation,total_time)
    
    # Plotting the best value per generation
    if plot_results:
        type_of_plot = int(input("\nPlease specify the type of result to display:\n*Type '1' for value increase per generation\n*Type '2' for encoding scheme per generation\n").strip())
        if type_of_plot == 1:
            plot_search_progress(best_individual_value_per_generation, type='value_per_gen')
        elif type_of_plot == 2:
            plot_search_progress(best_individual_per_generation, type='one_hot_encoding')
        
    # Return the final generation if asked
    if return_gen:
        return current_generation+1
    
    # Returning True is search was successful
    if best_individual_value == optimal_value:
        return True
    else:
        return False

# ================================================================== #
# ================================================================== #

# Choose the item package for the search
chosen_item_list, max_weight, optimal_value = large_items_package

# Set globals and hyperparameters for the evolutionary algorithm
best_individual = None
best_individual_value = 0
generations = 1000
progeny_per_generation = 250
parents_for_next_generation = 100
mutation_rate = 1

# ================================================================== #
# ================================================================== #

# Running main() if executed directly
if __name__ == "__main__":
    main()
```

## IV. Testing the accuracy and success rate of the evolutionary algorithm based search

```python
# Calculating the average success rate of the search
import matplotlib.pyplot as plt
number_of_searches = 100
successes = 0

# Choose the item package for the testing phase
chosen_item_list, max_weight, optimal_value = large_items_package

# Lists for capturing plotting variables
gen_numbers = []
failures = []

# Testing the algorithms iteratively
for i in range(number_of_searches):
    # Set globals and hyperparameters for the evolutionary algorithm
    best_individual = None
    best_individual_value = 0
    generations = 100
    progeny_per_generation = 500
    parents_for_next_generation = 250
    mutation_rate = 1
    
    # Using the 'return_gen' parameter to obtain generation values
    gen_num = main(print_report=False, plot_results=False, return_gen=True)
    
    # Appening the generations to the 'gen_numbers' list if the algorithm ended before max generations was reached
    if gen_num != generations:
        gen_numbers.append(gen_num)
        failures.append(i)
        successes += 1
        print(f"Search #{i+1}:\tSUCCESSFUL", end='\r')
    else: # Reaching max gen denotes a failure
        gen_numbers.append(0)
        failures.append(-1)
        print(f"Search #{i+1}:\tFAILED        ", end='\r')
        
# Printing the total success rate
print(f"\nTotal success rate: {successes}/{number_of_searches} ({successes/number_of_searches*100:.2f}%)")

# Generating x and y axes for plotting
xs = [x for x in range(len(gen_numbers))]
failed_gens = [-1 if x==0 else 0 for x in gen_numbers]

# Plotting the testing phase as a bar chart
plt.style.use("seaborn-darkgrid")
fig, ax = plt.subplots(figsize=(10,8), dpi=150)
ax.bar(xs,gen_numbers, label=f'Successful tests ({successes/number_of_searches*100:.2f}%)')
ax.bar(xs, failed_gens, color='red', label=f'Failed tests ({100 - (successes/number_of_searches*100):.2f}%)')

# Setting plot labels and legend
ax.set_title(f"Testing the evolutionary algorithm\n(Max generations={generations}, Recombining individuals={parents_for_next_generation},\nProgeny stength={progeny_per_generation}, Random mutation rate={mutation_rate})")
ax.set_ylabel("Generations created for optimal solution discovery")
ax.set_xlabel("Testing cycles", labelpad=10)
ax.legend(frameon=True, facecolor='white', shadow=True)
plt.show()
```
